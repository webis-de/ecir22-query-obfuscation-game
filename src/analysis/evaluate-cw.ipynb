```python
all_qrels = TrecQrel('topics-and-qrels/all-qrels.web.txt')

query_to_topic = {}

for trec in ['clueweb09/en/trec-web-2009', 'clueweb09/en/trec-web-2010', 'clueweb09/en/trec-web-2011', 'clueweb09/en/trec-web-2012', 'clueweb12/trec-web-2013', 'clueweb12/trec-web-2014']:
    for query in ir_datasets.load(trec).queries_iter():
        query_id = query.query_id
        query = query.query.lower().strip()
        if query in query_to_topic:
            raise ValueError('')
        query_to_topic[query] = int(query_id)

print(len(query_to_topic))
```

```python
def parse_trec_run(i, field_name=None):
    from trectools import TrecRun
    topic = query_to_topic[i['original query']]
    if topic <= 200:
        ranking = i['response ChatNoir cw09']
    else:
        ranking = i['response chat noir cw12']
    if field_name is not None:
        ranking = [j for j in pyserini_ranking(i['user query'])]
    
    trec_ranking = []
    for pos, entry in zip(range(len(ranking)), ranking):
        trec_ranking += [{
            'query': topic,
            'q0': 0,
            'docid': entry['trec id'],
            'rank': pos,
            'score': entry['score'],
            'system': 'does-not-matter'
        }]
    
    ret = TrecRun()
    ret.run_data = pd.DataFrame(trec_ranking)
    return ret

def pyserini_ranking(query):
    hits = SEARCHER.search(query, 5000)
    for hit in hits:
        yield {'trec id': uuid_to_trec_id[hit.docid], 'score': hit.score}

def qrels(i):
    from trectools import TrecQrel
    topic = query_to_topic[i['original query']]
    
    ret = TrecQrel()
    ret.qrels_data = all_qrels.qrels_data[all_qrels.qrels_data['query'].astype(int) == int(topic)].copy()
    
    return ret

def eval_data(run, qrels, topic):
    from trectools import TrecEval
    trec_eval = TrecEval(run, qrels)
    
    if len(run.run_data) < 1:
        return None
    return {'mrr': trec_eval.get_reciprocal_rank(depth=5000), 'ndcg_1000': trec_eval.get_ndcg(depth=1000), 'topic': topic, 'relevant': trec_eval.get_relevant_retrieved_documents()}
```

```python
cw_queries = []

# row 0
for _, i in tqdm(df[df['user query without stopwords'].str.len() == df['number of used keywords']].iterrows()):
    if i['original query'] in query_to_topic:
        topic = query_to_topic[i['original query']]
        run = parse_trec_run(i)
        run_qrels = qrels(i)
        cw_queries += [eval_data(run, run_qrels, topic)]
cw_queries = pd.DataFrame([i for i in cw_queries if i is not None])
cw_queries.describe()
```
