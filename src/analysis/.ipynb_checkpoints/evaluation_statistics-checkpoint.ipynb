```python
    # This plots the number of queries submitted by users seperated by levels
    # and the different categories
    import pandas as pd
    from matplotlib import pyplot as plt
    
    df = pd.read_json("prepared_log_data_evaluation_1.2.json")
    # shorten the data to the two relevent columns
    df = df[["category","level"]]
    tmp_df = df.loc[df["level"] == "squid"]
    
    # compute the frequence of the level squid for every category
    grouped_level_df["level squid"] = tmp_df.groupby("category").count()["level"]
    tmp_df = df.loc[df["level"] == "chameleon"]
    grouped_level_df["level chameleon"] = tmp_df.groupby("category").count()["level"]
    # delete unnecessary data
    grouped_level_df = grouped_level_df[["level squid", "level chameleon"]]
    # print(grouped_level_df)
    #ax = grouped_level_df.plot.bar()
    ax = grouped_level_df.plot(kind='bar')
    plt.title('Number of submited Queries per Category grouped by level')
    ax.set_ylabel('number of submited queries')
```

```python
    # Create a datastructure that evaluates different aspects of the following cases
    # 1. all words of the query are part of the keyword list
    # 2. at least one word of the list is part of the query AND at least one word is not from the list
    # 3. no word of the query is from the list
    # 4. there is at least one word that is not from the web page
    # For these cases we research the time needed to formulate a query, the mean reciprocal rank of the index and chatnoir,
    # the number of all queries that fit this criterium, the number of related documents and their average position
    # and the length of the queries
    
    from nltk.corpus import stopwords
    import pandas as pd
    import matplotlib.pyplot as plt
    
    stop = stopwords.words('english')
    pd.set_option("display.precision", 10)
    
    df = pd.read_json("prepared_log_data_evaluation_1.2.json")
    df = df[["document position in index", "document position in ChatNoir cw12", "number of used keywords", "used keywords",
            "number of used keywords from webpage", "used keywords from webpage", "user query is phrase from webpage", "user query",
            "last position in index", "last position ChatNoir cw12", "number of retrieved related documents index",
            "number of retrieved related documents cw12", "Time to enter Query"]]
    df = df[df["document position in ChatNoir cw12"].notna()]
    df['user query without stopwords'] = df['user query'].apply(lambda x: [word for word in x.split() if word.lower() not in (stop)])
    df['user query as string array'] = df['user query'].apply(lambda x: [word for word in x.split()])

    new_df = pd.DataFrame({"Mean Reciprocal Rank ChatNoir cw12": [0], "Mean Reciprocal Rank Index": [0],
                          "Number of Queries": [0], "Average Number of Related Documents ChatNoir cw12": [0],
                          "Average Number of Related Documents Index": [0],
                          "Average Time to formulate Query in Seconds": [0],
                          "Average Query Length": [0]})
    df = df[df["document position in ChatNoir cw12"].notna()]
    #print(df)
    # First row: All words of the query are from the list
    tmp_df = df[df['user query without stopwords'].str.len() == df['number of used keywords']]
    #print(tmp_df)
    new_df['Mean Reciprocal Rank Index'] = tmp_df['document position in index'].mean()
    new_df['Mean Reciprocal Rank ChatNoir cw12'] = tmp_df['document position in ChatNoir cw12'].mean()
    new_df['Number of Queries'] = len(tmp_df.index)
    new_df['Average Number of Related Documents ChatNoir cw12'] = tmp_df['number of retrieved related documents cw12'].mean()
    new_df['Average Number of Related Documents Index'] = tmp_df['number of retrieved related documents index'].mean()
    new_df['Average Time to formulate Query in Seconds'] = tmp_df['Time to enter Query'].mean()
    new_df['Average Query Length'] = tmp_df['user query as string array'].str.len().mean()
    
    # Second row: At least one word is from the list and at least one word is not from the list
    tmp_df = df[(df['number of used keywords'] != 0) &
                (df['user query without stopwords'].str.len() != df['number of used keywords'])]
    data = {}
    data['Mean Reciprocal Rank Index'] = tmp_df['document position in index'].mean()
    data['Mean Reciprocal Rank ChatNoir cw12'] = tmp_df['document position in ChatNoir cw12'].mean()
    data['Number of Queries'] = len(tmp_df.index)
    data['Average Number of Related Documents ChatNoir cw12'] = tmp_df['number of retrieved related documents cw12'].mean()
    data['Average Number of Related Documents Index'] = tmp_df['number of retrieved related documents index'].mean()
    data['Average Time to formulate Query in Seconds'] = tmp_df['Time to enter Query'].mean()
    data['Average Query Length'] = tmp_df['user query as string array'].str.len().mean()

    new_df = new_df.append(data, ignore_index = True)
    
    # Third row: No word of the query is from the list
    tmp_df = df[df['number of used keywords'] == 0]
    data['Mean Reciprocal Rank Index'] = tmp_df['document position in index'].mean()
    data['Mean Reciprocal Rank ChatNoir cw12'] = tmp_df['document position in ChatNoir cw12'].mean()
    data['Number of Queries'] = len(tmp_df.index)
    data['Average Number of Related Documents ChatNoir cw12'] = tmp_df['number of retrieved related documents cw12'].mean()
    data['Average Number of Related Documents Index'] = tmp_df['number of retrieved related documents index'].mean()
    data['Average Time to formulate Query in Seconds'] = tmp_df['Time to enter Query'].mean()
    data['Average Query Length'] = tmp_df['user query as string array'].str.len().mean()
    
    new_df = new_df.append(data, ignore_index = True)
    
    # Forth row: At least one word of the query is not from the target document
    tmp_df = df[df['user query without stopwords'].str.len() > df['number of used keywords from webpage']]
    #print(tmp_df)
    data['Mean Reciprocal Rank Index'] = tmp_df['document position in index'].mean()
    data['Mean Reciprocal Rank ChatNoir cw12'] = tmp_df['document position in ChatNoir cw12'].mean()
    data['Number of Queries'] = len(tmp_df.index)
    data['Average Number of Related Documents ChatNoir cw12'] = tmp_df['number of retrieved related documents cw12'].mean()
    data['Average Number of Related Documents Index'] = tmp_df['number of retrieved related documents index'].mean()
    data['Average Time to formulate Query in Seconds'] = tmp_df['Time to enter Query'].mean()
    data['Average Query Length'] = tmp_df['user query as string array'].str.len().mean()
    
    new_df = new_df.append(data, ignore_index = True)
    
    print(new_df)
    #print(tmp_df)
    
    #ax = new_df.plot(kind='bar')
```
